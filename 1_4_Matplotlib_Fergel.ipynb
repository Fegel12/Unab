## Pyplot

En Matplotlib, [pyplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html) es un módulo con funciones para crear gráficos de forma sencilla e interactiva.


*   **Figure** (fig) es un objeto que actúa como el contenedor principal de todos los elementos de una gráfica (como subtramas y ejes).
*   **Axes** (ax) es un objeto que representa un único (sub)gráfico dentro de la Figure, conteniendo los datos, etiquetas y el título de ese gráfico específico.

```python
fig, ax = plt.subplots()
plt.show()
```

Veamos una tabla cruzada generada por el puntaje crediticio agrupado por edad.

```python
df_by_age =df.groupby('Age')['CreditScore'].agg([min, max, np.mean])
```

```python
df_by_age.head()
```

```python
df_by_age.index
```

Podemos ver varios graficos en el mismo entorno ax

```python
fig, ax = plt.subplots()
ax.plot(df_by_age.index, df_by_age["mean"]) #ax.plot(ejex, ejey)
ax.plot(df_by_age.index, df_by_age["min"])
ax.plot(df_by_age.index, df_by_age["max"])
plt.show()
```

Esta gráfica se puede customizar

```python
fig, ax = plt.subplots()
# gráficas
ax.plot(df_by_age.index, df_by_age["mean"], marker='o' , linestyle="-", color='b')
ax.plot(df_by_age.index, df_by_age["min"], marker='x', linestyle="--", color='r')
ax.plot(df_by_age.index, df_by_age["max"], marker='v', linestyle='None', color='gray')
# nombres de los ejes
ax.set_xlabel("Age")
ax.set_ylabel("Puntaje Crediticio")
#titulo
ax.set_title("Comparación puntajes por edad")
#leyenda
plt.legend(["Mean","Min", "Max"])
plt.show()
```

También podemos separar los gráficos, por ejemplo 3 filas con 2 columnas

```python
fig, ax = plt.subplots(3, 2)
plt.show()
```

En nuestro caso 3 filas y 1 columna sería así:

```python
fig, ax = plt.subplots(3, 1)
#primer gráfico
ax[0].plot(df_by_age.index, df_by_age["mean"], color='b')
ax[0].set_xlabel("Age")
ax[0].set_ylabel("Media")
#segundo gráfico
ax[1].plot(df_by_age.index, df_by_age["min"],  color='r')
ax[1].set_xlabel("Age")
ax[1].set_ylabel("Minimo")
#segundo gráfico
ax[2].plot(df_by_age.index, df_by_age["max"],  color='g')
ax[2].set_xlabel("Age")
ax[2].set_ylabel("Max")
plt.show()
```

Creemos otra variable para hacer un $2 \times 2$

```python
df_by_age['otro'] = df_by_age['max'] - df_by_age['min']
```

```python
fig, ax = plt.subplots(2, 2)
#primer gráfico
ax[0,0].plot(df_by_age.index, df_by_age["mean"], color='b')
ax[0,0].set_xlabel("Age")
ax[0,0].set_ylabel("Media")
#segundo gráfico
ax[1,0].plot(df_by_age.index, df_by_age["min"],  color='r')
ax[1,0].set_xlabel("Age")
ax[1,0].set_ylabel("Minimo")
#segundo gráfico
ax[0,1].plot(df_by_age.index, df_by_age["max"],  color='g')
ax[0,1].set_xlabel("Age")
ax[0,1].set_ylabel("Max")
#Otro gráfico agregado
ax[1,1].plot(df_by_age.index, df_by_age["otro"],  color='gray')
ax[1,1].set_xlabel("Age")
ax[1,1].set_ylabel("Semi-suma")
plt.show()
```

## Diagramas de Barras

Hagamos una tabla pivote por edad (filas) y Tenure (columnas) promediando los puntajes crediticios

```python
df_pivot = df.pivot_table(index='Age', columns='Geography', values='CreditScore', aggfunc='mean', fill_value=0)
df_pivot.head()
```

```python
fig,ax =plt.subplots()
ax.bar(df_pivot.index, df_pivot['Germany'])
plt.show()
```

Podemos hacer barras acumuladas:

```python
fig,ax =plt.subplots()
ax.bar(df_pivot.index, df_pivot['Germany'])
ax.bar(df_pivot.index, df_pivot['Spain'], bottom=df_pivot['Germany'])
#ax.bar(df_pivot.index, df_pivot[2], bottom=df_pivot[0] + df_pivot[1])
#ax.bar(df_pivot.index, df_pivot[3], bottom=df_pivot[0] + df_pivot[1] + df_pivot[2])
ax.legend(['Germany', 'Spain'])#, 'Tenure 2', 'Tenure 3'])
plt.show()
```

Por ejemplo en el gráfico de medias podemos usar indicadores de error

```python
fig, ax = plt.subplots()
ax.bar('Female', df[df['Gender']=='Female']['CreditScore'].mean(), yerr= df[df['Gender']=='Female']['CreditScore'].std())
ax.bar('Male', df[df['Gender']=='Male']['CreditScore'].mean(), yerr= df[df['Gender']=='Male']['CreditScore'].std())
plt.show()
```

O hacer histogramas

```python
fig, ax = plt.subplots()
ax.hist(df[df['Gender']=='Female']['CreditScore'], bins=15, histtype="step")
ax.hist(df[df['Gender']=='Male']['CreditScore'], bins=15, histtype="step")
ax.legend(['Female', 'Male'])
ax.set_xlabel("Puntaje")
ax.set_ylabel("# of observations")
plt.show()
```

## Boxplots y scatterplots

```python
df.head()
```

Veamos primero un par de gráficos de cajas:

```python
fig, ax = plt.subplots()
ax.boxplot([
    df[df['Gender']=='Female']['CreditScore'],
    df[df['Gender']=='Male']['CreditScore']
    ])
ax.set_xticklabels(['Female', 'Male'])
ax.set_ylabel("Score")
ax.set_title('Comparación de puntaje por Género')
plt.show()
```

```python
fig,ax =plt.subplots()
ax.boxplot([ df_pivot['France'],
        df_pivot['Germany'],
       df_pivot['Spain']])
ax.set_xticklabels(['France', 'Germany', 'Spain'])
ax.set_title('Comparación de puntaje por País')
plt.show()
```

Veamos la distribución de Balance vs Edad:

```python
fig, ax = plt.subplots()
ax.scatter(
    df['Age'],
    df['Balance']
    )
ax.set_ylabel("Balance")
ax.set_xlabel("Edad")
ax.set_title('Comparación de Balance por Edad')
plt.show()
```

podemos separarlos por género

```python
fig, ax = plt.subplots()
ax.scatter(
    df[df['Gender']=='Female']['Age'],
    df[df['Gender']=='Female']['Balance'],
    color='Blue',
    label='Female'
    )
ax.scatter(
    df[df['Gender']=='Male']['Age'],
    df[df['Gender']=='Male']['Balance'],
    color='Red',
    label='Male'
    )
ax.legend()
ax.set_ylabel("Balance")
ax.set_xlabel("Edad")
ax.set_title('Comparación de Balance por Edad')
plt.show()
```

Podemos agregarle una tercera variable numérica para los colores (muy bueno cuando hay clusters)

```python
fig, ax = plt.subplots()
ax.scatter(
    df['Age'],
    df['Balance'],
    c=df.Tenure
    )
ax.set_ylabel("Balance")
ax.set_xlabel("Edad")
ax.set_title('Comparación de Balance por Edad')
plt.show()
```

## Estilo del gráfico

```python
plt.style.use('ggplot')
fig, ax = plt.subplots()
ax.scatter(
    df['Age'],
    df['Balance'],
    c=df.Tenure
    )
ax.set_ylabel("Balance")
ax.set_xlabel("Edad")
ax.set_title('Comparación de Balance por Edad')
plt.show()
```

```python
plt.style.use('bmh')
fig, ax = plt.subplots()
ax.scatter(
    df['Age'],
    df['Balance'],
    c=df.Tenure
    )
ax.set_ylabel("Balance")
ax.set_xlabel("Edad")
ax.set_title('Comparación de Balance por Edad')
plt.show()
```

También se puede ajustar el tamaño del gráfico

```python
fig, ax = plt.subplots()
fig.set_size_inches([5, 3])
ax.scatter(
    df['Age'],
    df['Balance'],
    c=df.Tenure
    )
ax.set_ylabel("Balance")
ax.set_xlabel("Edad")
ax.set_title('Comparación de Balance por Edad')
plt.show()
```
