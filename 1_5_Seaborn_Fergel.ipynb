# Seaborn

**Seaborn** es una biblioteca de visualización de datos basada en Matplotlib que proporciona una interfaz de alto nivel para crear gráficos estadísticos atractivos y más fáciles de interpretar. Seaborn está especialmente diseñada para trabajar con pandas DataFrames y facilita la creación de gráficos complejos con menos código que Matplotlib. Ofrece funciones preconfiguradas para gráficos como distribuciones, gráficos de correlación, gráficos categóricos, entre otros.

La documentación está disponible en [link](https://seaborn.pydata.org/)

**Ventajas:**

* Fácil de usar
* Funciona bien con las estructuras de datos de Pandas
* Desarrollado sobre matplotlib

## Otros gráficos

Otros gráficos bastante usados son los siguientes:
- kdeplot (kernel density estimate (KDE) plot) es un método para visualizar la distribución de observaciones en un conjunto de datos, similar a un histograma. KDE representa los datos mediante una curva de densidad de probabilidad continua en una o más dimensiones.

```python
sns.kdeplot(x='Fat_Percentage', data=df)
plt.show()
```

- Un ECDF representa la proporción o el recuento de observaciones que caen por debajo de cada valor único en un conjunto de datos. En comparación con un histograma o un gráfico de densidad, tiene la ventaja de que cada observación se visualiza directamente, lo que significa que no es necesario ajustar parámetros de binning o suavizado. También facilita las comparaciones directas entre múltiples distribuciones. Una desventaja es que la relación entre la apariencia del gráfico y las propiedades básicas de la distribución (como su tendencia central, varianza y la presencia de cualquier bimodalidad) puede no ser tan intuitiva.

```python
sns.ecdfplot(x='Fat_Percentage', data=df)
plt.show()
```

- swarmplot es un diagrama de dispersión categórico con puntos ajustados para que no se superpongan.

```python
sns.swarmplot(x='Fat_Percentage', y='Workout_Type', data=df)
plt.show()
```

- Un diagrama de violín cumple una función similar a un diagrama de caja y bigotes. Muestra la distribución de los puntos de datos tras agruparlos por una (o más) variables.

```python
sns.violinplot(y='Workout_Type', x='Fat_Percentage',data=df)
plt.show()
```

Para ver mas opciones se recomienda ver la webpage de [Seaborn](https://seaborn.pydata.org/index.html)

## Relplot

Crear "gráficos relacionales": gráficos de dispersión o gráficos de líneas

- ¿Por qué usar relplot() en lugar de scatterplot()?

  relplot() permite crear subgráficos en una sola figura.

Se pueden asociar en columnas

```python
sns.relplot(x='Height',
            y='Weight',
            data=df,
            kind='scatter',
            col='Gender')
plt.show()
```

Se pueden asociar en filas

```python
sns.relplot(x='Height',
            y='Weight',
            data=df,
            kind='scatter',
            row='Gender')
plt.show()
```

Se puede hacer una matriz de scatterplots

```python
sns.relplot(x='Height',
            y='Weight',
            data=df,
            kind='scatter',
            row='Gender',
            col='Workout_Type')
plt.show()
```

O se puede asociar convenientemente

```python
sns.relplot(x='Height',
            y='Weight',
            data=df,
            kind='scatter',
            col='Workout_Type',
            col_wrap=2)
plt.show()
```

También se pueden hacer ajustes de grupo o tamaño del punto:

```python
sns.relplot(x='Height',
            y='Weight',
            data=df,
            kind='scatter',
            col='Workout_Type',
            col_wrap=2,
            size='Workout_Frequency',
            hue='Experience_Level')
plt.show()
```

Se pueden usar Lineplots en este caso:

Para ajustar los datos a un caso en donde se usaría este tipo de gráfico voy a usar un groupby en la altura (filtrando para hombres):

```python
filter_man= df[df['Gender']=='Male'].groupby('Height')['Weight'].agg([min, max, np.mean, np.std])
```

```python
filter_man.head()
```

```python
sns.relplot(x=filter_man.index,
            y='mean',
            data=filter_man,
            kind='line')
plt.show()
```

Si no hicieramos esto, sería así:

```python
sns.relplot(x='Height',
            y='Weight',
            data=df[df['Gender']=='Male'],
            kind='line')
plt.show()
```

Podemos hacer que el rango sea la desviación estándar

```python
sns.relplot(x='Height',
            y='Weight',
            data=df[df['Gender']=='Male'],
            kind='line',
            errorbar=None)
plt.show()
```

Observación: usando 'None' sería el mismo gráfico inicial.

## Catplot

Se utiliza para crear gráficos categóricos, con las mismas ventajas que relplot().

- Crea fácilmente subgráficos con col= y row=.

Para mas info ver [Catplot](https://seaborn.pydata.org/generated/seaborn.catplot.html)

```python
sns.catplot(x='Workout_Type',
            data=df,
            kind='count',
            col='Gender')
plt.show()
```

Pero se puede usar con variables numéricas en el eje y (en ese caso se usa tipo bar, y se gráfica la media)

```python
sns.catplot(x='Workout_Type',
            y='Session_Duration',
            data=df,
            kind='bar',
            col='Gender')
plt.show()
```

La linea sobre las barras es el CI 95% (('ci', 95)), si no se quiere se debe hacer lo siguiente:

```python
sns.catplot(x='Workout_Type',
            y='Session_Duration',
            data=df,
            kind='bar',
            col='Gender',
            errorbar=None)
plt.show()
```

Cambiando el orden de los ejes se cambia la orientación

```python
sns.catplot(y='Workout_Type',
            x='Session_Duration',
            data=df,
            kind='bar',
            col='Gender',
            errorbar=None)
plt.show()
```

También se puede hacer boxplot con esta característica, sus ventajas son:
- Muestra la distribución de datos cuantitativos.
- Ver mediana, dispersión, asimetría y valores atípicos.
- Facilita las comparaciones entre grupos.

```python
sns.catplot(x='Workout_Type',
            y='Fat_Percentage',
            data=df,
            kind='box',
            col='Gender'
            )
plt.show()
```

Hay opciones similares como:
- Violinplot: Usar kind="violin"
- Bodnplot: kind="boxen".

Otra opción es el diagrama de puntos:

- Los puntos muestran la media de la variable cuantitativa.

- Las líneas verticales muestran intervalos de confianza del 95 %.

```python
sns.catplot(x='Workout_Type',
            y='Fat_Percentage',
            data=df,
            kind='point',
            hue='Gender'
            )
plt.show()
```

Se pueden quitar las lineas que los unen

```python
sns.catplot(x='Workout_Type',
            y='Fat_Percentage',
            data=df,
            kind='point',
            hue='Gender',
            linestyle='None'
            )
plt.show()
```

Se puede usar la mediana como estimador:

```python
sns.catplot(x='Workout_Type',
            y='Fat_Percentage',
            data=df,
            kind='point',
            hue='Gender',
            estimator=np.median
            )
plt.show()
```

Se puede quitar el intervalo de confianza

```python
sns.catplot(x='Workout_Type',
            y='Fat_Percentage',
            data=df,
            kind='point',
            hue='Gender',
            estimator=np.median,
            ci=None
            )
plt.show()
```

## Titulos y labels

Los gráficos de Seaborn crean dos tipos diferentes de objetos:
- FacetGrid: Se usa en relplot() y catplot(), Puede crear subplots.
- AxesSubplot: Se usa en scatterplot() , countplot() , etc. Solo crea una única gráfica.

```python
g= sns.catplot(x='Workout_Type',
            y='Fat_Percentage',
            data=df,
            kind='box',
            col='Gender'
            )
g.fig.suptitle('Fat Percentage by Workout Type',
               y=1.2 #altura del titulo
               )
g.set_titles('Gráfico de {col_name}')
plt.show()
```

```python
g= sns.boxplot(x='Workout_Type',
            y='Fat_Percentage',
            data=df,
            hue='Gender'
            )
g.set_title('Fat Percentage by Workout Type',
               y=1.2 #altura del titulo
               )
g.set(xlabel='Workout Type', ylabel='Fat Percentage')
plt.legend(loc='upper left')
plt.show()
```
